```python
#import libs
import pandas as pd
#import module glob fo create file lists to join'
import glob
```

```python
# all csv files are interesing for us so * means all csv files
files = glob.glob('*.csv')
#make view all csv files
files
#see just
#[]
#couse no data
```

```python
#we join all files in one
ttl_files = pd.DataFrame()
#leat craete list
for file in files:
  #creare new data and read all files
  data = pd.read_csv(file)
  #create new colums
  #переменная файл находися в каждом элементе списка
  data['common_file'] =file
  #переназначим переменную  ttl_files
  ttl_files = pd.concat([ttl_files, data])
```

```python
# see what we done
ttl_files
```

```python
#create now common .csv file
ttl_files.to_csv('ttl_files.csv', index=False)
```
